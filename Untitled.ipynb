```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sphere(r):

    u = np.linspace(0, 2 * np.pi, r)

    v = np.linspace(0, np.pi, r)

    sx = r * np.outer(np.cos(u), np.sin(v))

    sy = r * np.outer(np.sin(u), np.sin(v))

    sz = r * np.outer(np.ones(np.size(u)), np.cos(v))

    return sx,sy,sz
```

```python
def sphere2(r,d):

    u = np.linspace(0, 2 * np.pi, r)

    v = np.linspace(0, np.pi, r)

    sx = r * np.outer(np.cos(u), np.sin(v))+d

    sy = r * np.outer(np.sin(u), np.sin(v))

    sz = r * np.outer(np.ones(np.size(u)), np.cos(v))

    return sx,sy,sz
```
