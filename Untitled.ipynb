```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import gammainc
```

```python
def sphere(r):

    u = np.linspace(0, 2 * np.pi, r)

    v = np.linspace(0, np.pi, r)

    sx = r/2 * np.outer(np.cos(u), np.sin(v))

    sy = r/2 * np.outer(np.sin(u), np.sin(v))

    sz = r/2 * np.outer(np.ones(np.size(u)), np.cos(v))

    return sx,sy,sz
```

```python
def sphere2(r,d):

    u = np.linspace(0, 2 * np.pi, r)

    v = np.linspace(0, np.pi, r)

    sx = r/2 * np.outer(np.cos(u), np.sin(v))+d

    sy = r/2 * np.outer(np.sin(u), np.sin(v))

    sz = r/2 * np.outer(np.ones(np.size(u)), np.cos(v))

    return sx,sy,sz
```

```python
def new_positions_spherical_coordinates(self):
   radius = np.random.uniform(0.0,1.0, (self.number_of_particles,1)) 
   theta = np.random.uniform(0.,1.,(self.number_of_particles,1))*pi
   phi = np.arccos(1-2*np.random.uniform(0.0,1.,(self.number_of_particles,1)))
   x = radius * np.sin( theta ) * np.cos( phi )
   y = radius * np.sin( theta ) * np.sin( phi )
   z = radius * np.cos( theta )
   return (x,y,z)
```

```python
def sample(center,radius,n_per_sphere):
    r = radius
    ndim = center.size
    x = np.random.normal(size=(n_per_sphere, ndim))
    ssq = np.sum(x**2,axis=1)
    fr = r*gammainc(ndim/2,ssq/2)**(1/ndim)/np.sqrt(ssq)
    frtiled = np.tile(fr.reshape(n_per_sphere,1),(1,ndim))
    p = center + np.multiply(x,frtiled)
    return p

fig1 = plt.figure(1)
ax1 = fig1.gca()
center = np.array([0,0])
radius = 1
p = sample(center,radius,204)
ax1.scatter(p[:,0],p[:,1],s=0.5)
ax1.add_artist(plt.Circle(center,radius,fill=False,color='0.5'))
ax1.set_xlim(-1.5,1.5)
ax1.set_ylim(-1.5,1.5)
ax1.set_aspect('equal')
```
