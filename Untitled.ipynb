```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sphere(r):

    u = np.linspace(0, 2 * np.pi, r)

    v = np.linspace(0, np.pi, r)

    sx = r * np.outer(np.cos(u), np.sin(v))

    sy = r * np.outer(np.sin(u), np.sin(v))

    sz = r * np.outer(np.ones(np.size(u)), np.cos(v))

    return sx,sy,sz
```

```python
sx1,sy1,sz1 = sphere(10)
sx2,sy2,sz2 = sphere(9)
sx3,sy3,sz3 = sphere(8)
sx4,sy4,sz4 = sphere(7)
sx5,sy5,sz5 = sphere(6)
sx6,sy6,sz6 = sphere(5)
sx7,sy7,sz7 = sphere(4)
sx8,sy8,sz8 = sphere(3)
sx9,sy9,sz9 = sphere(2)
sx10,sy10,sz10 = sphere(1)
x=np.append(sx1,sx2)
x=np.append(x,sx3)
x=np.append(x,sx4)
x=np.append(x,sx5)
x=np.append(x,sx6)
x=np.append(x,sx7)
x=np.append(x,sx8)
x=np.append(x,sx9)
x=np.append(x,sx10)
y=np.append(sy1,sy2)
y=np.append(y,sy3)
y=np.append(y,sy4)
y=np.append(y,sy5)
y=np.append(y,sy6)
y=np.append(y,sy7)
y=np.append(y,sy8)
y=np.append(y,sy9)
y=np.append(y,sy10)
z=np.append(sz1,sz2)
z=np.append(z,sz3)
z=np.append(z,sz4)
z=np.append(z,sz5)
z=np.append(z,sz6)
z=np.append(z,sz7)
z=np.append(z,sz8)
z=np.append(z,sz9)
z=np.append(z,sz10)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x, y, z, color='r')
plt.show()
```
